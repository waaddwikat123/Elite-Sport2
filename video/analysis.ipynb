```python
data = {
    "faculty": np.random.choice(["Engineering", "Arts"], 100),

    "gender": np.random.choice(["Male", "Female"], 100),

    "screen_hours": np.random.choice(
        [
            "Less than 2 hours",
            "2 - 4 hours",
            "4 - 6 hours",
            "6 - 8 hours",
            "More than 8 hours"
        ],
        100
    ),

    # Study Questions
    "study_daily": np.random.randint(1, 6, 100),
    "study_long_sessions": np.random.randint(1, 6, 100),
    "study_evening": np.random.randint(1, 6, 100),
    "study_before_sleep": np.random.randint(1, 6, 100),

    # Entertainment Questions
    "ent_daily": np.random.randint(1, 6, 100),
    "ent_long_sessions": np.random.randint(1, 6, 100),
    "social_media_usage": np.random.randint(1, 6, 100),
    "phone_in_bed": np.random.randint(1, 6, 100),

    # Sleep Questions
    "sleep_difficulty": np.random.randint(1, 6, 100),
    "long_time_to_sleep": np.random.randint(1, 6, 100),
    "wake_up_night": np.random.randint(1, 6, 100),
    "comfortable_sleep": np.random.randint(1, 6, 100),
    "wake_up_fresh": np.random.randint(1, 6, 100),

    # Actual Sleep Data
    "bedtime": np.random.choice(
        [
            "Before 10 PM",
            "10 PM - 12 AM",
            "12 AM - 2 AM",
            "After 2 AM"
        ],
        100
    ),

    "sleep_duration": np.random.choice(
        [
            "Less than 6 hours",
            "6 - 9 hours",
            "More than 9 hours"
        ],
        100
    )
}

df = pd.DataFrame(data)

df.head()
```

8) Creating Composite Scores

Several related survey questions were combined into one average score called a Composite Score.

This was done to simplify the dataset and represent each research dimension using one measurable variable instead of analyzing every question separately.

For example:
- study_daily
- study_long_sessions
- study_evening
- study_before_sleep

were combined into:

study_usage_score

using the average of the four questions.

تم دمج عدة أسئلة مرتبطة بنفس البُعد داخل متغير واحد باستخدام المتوسط الحسابي، وذلك لإنشاء Composite Scores تمثل:

الاستخدام الدراسي
الاستخدام الترفيهي
جودة النوم

```python
df["study_usage_score"] = df[
    [
        "study_daily",
        "study_long_sessions",
        "study_evening",
        "study_before_sleep"
    ]
].mean(axis=1)

df["entertainment_usage_score"] = df[
    [
        "ent_daily",
        "ent_long_sessions",
        "social_media_usage",
        "phone_in_bed"
    ]
].mean(axis=1)

df["sleep_quality_score"] = df[
    [
        "sleep_difficulty",
        "long_time_to_sleep",
        "wake_up_night"
    ]
].mean(axis=1)

df.head()
```

9) Correlation Analysis

Correlation is a statistical method used to measure the relationship between two variables.

It helps determine:
- whether a relationship exists,
- the direction of the relationship,
- and the strength of the relationship.

In this project, correlation was used to study the relationship between:
- study usage,
- entertainment usage,
- and sleep quality.

```python
correlation = df[
    [
        "study_usage_score",
        "entertainment_usage_score",
        "sleep_quality_score"
    ]
].corr()

correlation
```

11) Calculating Correlation

We used the following command:

corr()

The .corr() function calculates the correlation coefficient between variables and displays the results in a Correlation Matrix.


12) Understanding Correlation Values

Correlation values range from:

-1 to +1

Interpretation:

• Close to +1:
Strong positive relationship

• Close to -1:
Strong negative relationship

• Close to 0:
Weak or no relationship

13) Scatter Plot

A Scatter Plot was used to visualize the relationship between:
- entertainment usage
- and sleep quality

Each point on the graph represents one participant.

The graph helps identify:

```python
plt.scatter(
    df["entertainment_usage_score"],
    df["sleep_quality_score"]
)

plt.xlabel("Entertainment Usage Score")
plt.ylabel("Sleep Quality Score")
plt.title("Entertainment Usage vs Sleep Quality")

plt.show()
```

يمثل كل نقطة مشاركًا واحدًا في الدراسة، ويُستخدم الرسم لفحص العلاقة البصرية بين الاستخدام الترفيهي للشاشات وجودة النوم. وبما أن البيانات الحالية عشوائية، لا يظهر اتجاه قوي واضح بين المتغيرين.

```python
screen_hours_numeric = {
    "Less than 2 hours": 1,
    "2 - 4 hours": 3,
    "4 - 6 hours": 5,
    "6 - 8 hours": 7,
    "More than 8 hours": 9
}
sleep_duration_numeric = {
    "Less than 6 hours": 5,
    "6 - 9 hours": 7,
    "More than 9 hours": 10
}
df["screen_hours_num"] = df["screen_hours"].map(screen_hours_numeric)

df["sleep_duration_num"] = df["sleep_duration"].map(sleep_duration_numeric)
df[["screen_hours_num", "sleep_duration_num"]].corr()
```

```python
engineering_sleep = df[df["faculty"] == "Engineering"]["sleep_quality_score"]

arts_sleep = df[df["faculty"] == "Arts"]["sleep_quality_score"]
```

Important Note

The dataset used in this project was randomly generated using NumPy.

Therefore:

the T-Test result is only for demonstration purposes and does not represent real scientific findings.

```python
t_stat, p_value = ttest_ind(engineering_sleep, arts_sleep)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```
